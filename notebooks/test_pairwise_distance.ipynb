```python
from rdkit import Chem, DataStructs
from rdkit.Chem import AllChem
import numpy as np
from tqdm import tqdm  # For progress tracking
import time

def compute_similarity_matrix(smiles_list, similarity_metric=DataStructs.TanimotoSimilarity, 
                              fingerprint_type="morgan", radius=2, nBits=2048, 
                              return_distance=True):
    """
    Compute a similarity/distance matrix for a list of SMILES strings.
    
    Parameters:
    -----------
    smiles_list : list
        List of SMILES strings
    similarity_metric : function
        Similarity metric function from DataStructs (default: TanimotoSimilarity)
    fingerprint_type : str
        Type of fingerprint to use: "morgan" or "rdkit"
    radius : int
        Radius for Morgan fingerprint (default: 2)
    nBits : int
        Number of bits in the fingerprint (default: 2048)
    return_distance : bool
        If True, return distance matrix (1-similarity), else return similarity matrix
        
    Returns:
    --------
    numpy.ndarray
        Similarity or distance matrix
    """
    # Convert SMILES to RDKit molecules
    mols = []
    valid_indices = []
    
    for i, smiles in enumerate(smiles_list):
        mol = Chem.MolFromSmiles(smiles)
        if mol is not None:
            mols.append(mol)
            valid_indices.append(i)
    
    # Generate fingerprints
    fps = []
    for mol in mols:
        if fingerprint_type.lower() == "morgan":
            fp = AllChem.GetMorganFingerprintAsBitVect(mol, radius=radius, nBits=nBits)
        elif fingerprint_type.lower() == "rdkit":
            fp_gen = AllChem.GetRDKitFPGenerator(maxPath=7, fpSize=nBits)
            fp = fp_gen.GetFingerprint(mol)
        else:
            raise ValueError(f"Unsupported fingerprint type: {fingerprint_type}")
        fps.append(fp)
    
    # Calculate similarity matrix
    n = len(fps)
    similarity_matrix = np.zeros((len(smiles_list), len(smiles_list)))
    
    # Compute upper triangle
    for i in tqdm(range(n)):
        similarity_matrix[valid_indices[i], valid_indices[i]] = 1.0  # Self-similarity
        for j in range(i+1, n):
            sim = similarity_metric(fps[i], fps[j])
            similarity_matrix[valid_indices[i], valid_indices[j]] = sim
            similarity_matrix[valid_indices[j], valid_indices[i]] = sim  # Matrix is symmetric
    
    if return_distance:
        return 1 - similarity_matrix
    else:
        return similarity_matrix

# Compute distance matrix using Tanimoto coefficient
tic = time.time()
distance_matrix = compute_similarity_matrix(smiles, similarity_metric=DataStructs.TanimotoSimilarity)
print(time.time() - tic)
print(distance_matrix)

# You can also compute using other metrics
# similarity_matrix = compute_similarity_matrix(smiles, similarity_metric=DataStructs.DiceSimilarity, return_distance=False)
```
