```python
# Add columns to check validity and get standardized SMILES
def validate_and_standardize_smiles(smiles):
    try:
        mol = Chem.MolFromSmiles(smiles)
        if mol is None:
            return False, None
        return True, Chem.MolToSmiles(mol, isomericSmiles=True)
    except:
        return False, None

# Apply the validation to the SMILES column
valid_smiles = []
standard_smiles = []

for smile in data['SMILES']:
    is_valid, std_smiles = validate_and_standardize_smiles(smile)
    valid_smiles.append(is_valid)
    standard_smiles.append(std_smiles)

data['valid_smiles'] = valid_smiles
data['standard_smiles'] = standard_smiles

# Print summary statistics
print(f"Total SMILES: {len(data)}")
print(f"Valid SMILES: {sum(valid_smiles)}")
print(f"Invalid SMILES: {len(data) - sum(valid_smiles)}")

# Show a few examples of invalid SMILES if any exist
if len(data) - sum(valid_smiles) > 0:
    print("\nExamples of invalid SMILES:")
    print(data[~data['valid_smiles']][['SMILES']].head())
```
